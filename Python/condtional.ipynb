```python
amount=int(input("enter amount:"))
if amount<2000:
    print("No discount")
elif amount<=5000:
    print("you got 10% discount,amount after discount is",amount-(amount/10))
else:
    print("you got 20% discount,amount after discount is",amount-(amount/5))
print("Thanks for shopping")
```
