Encapsulation 
it is used to hide necessary data 

```python
class Company:
    company_name="Abc" #public 
    company_add="pilani" #public
    __amount=1234567 #private
comp=Company()
comp.company_name
# comp.amount
```

```python
# Accessing private variable 
class Company:
    company_name="Abc" 
    company_add="pilani" 
    __amount=1234567 
    def info(a):
        print(a.__amount)
    #or
    #comp._Company__amount

comp=Company()
comp.company_name
comp.info()
```
