```python
class Student:
    def display(self): # you have to pass a variable in method 
        name="Hitesh"
        classes="2nd yr"
        print(F"My name is {name} and I am in {classes}")
st=Student()
st.display()
```

```python
class Student1:
    name="Hitesh"
    classes="2nd yr"
    def display(a,b): 
        print(F"My name is {a.name} and I am in {a.classes}")
        print(b)
st1=Student1()
st1.display(b=10)
```

```python
class Company:
    company_name="Abc"
    company_add="pilani"

class Developer(Company):# single Inheritance
    developer_name="Hitesh"
    def info(a):
        print(a.company_name)
        print(a.company_add)
        print(a.developer_name)
dev=Developer()
dev.info()
```

```python
# multiple Inheritance
class Company:
    company_name="Abc"
    company_add="pilani"

class Client:
    cilent_name="xyz"
    pass
class Developer(Company,Client):
    developer_name="Hitesh"
    def info(a):
        print(a.company_name)
        print(a.company_add)
        print(a.developer_name)
        print(a.cilent_name)
dev=Developer()
dev.info()

```

```python
class Company:
    company_name="Abc"
    company_add="pilani"

class Client:
    cilent_name="xyz"
    
class Developer(Company,Client):
    developer_name="Hitesh"
    def info(a):
        print(a.company_name)
        print(a.company_add)
        print(a.developer_name)
        print(a.cilent_name)

class Project(Developer):
      project_name="xyz"
      def info(a):
        print(a.project_name)
        print(a.developer_name)
      
dev=Developer()
dev.info()
pr=Project()
pr.info()
```
